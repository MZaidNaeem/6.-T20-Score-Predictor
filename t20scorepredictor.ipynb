```python
import pandas as pd
import numpy as np
from yaml import safe_load
import os 
from tqdm  import tqdm
import warnings
import yaml
warnings.filterwarnings("ignore")
```

```python
filenames = []
for file in os.listdir('data'):
    filenames.append(os.path.join('data',file))
```

```python
final_df=pd.DataFrame()
counter=1
for i in tqdm(filenames):
    with open(i,'r') as f:
        data = yaml.safe_load(f)
        if isinstance(data, list) or isinstance(data, dict):
            df = pd.json_normalize(data)
        else:
            raise ValueError(f"File {i} does not contain a list or dict")
        df['match_id']=counter
        final_df=pd.concat([final_df,df],ignore_index=True)
        counter+=1
final_df
```

```python
def bowl(row):
    return ast.literal_eval(row['teams'])[1]
        
```

```python
data['bowling_team'] = data.apply(bowl,axis=1)
```

```python
data.drop(columns=['teams'],inplace=True)
```

```python
teams = [
    'Australia',
    'India',
    'Bangladesh',
    'New Zealand',
    'South Africa',
    'England',
    'West Indies',
    'Afghanistan',
    'Pakistan',
    'Sri Lanka'    
]
```

```python
data = data[data['batting_team'].isin(teams)]
data = data[data['bowling_team'].isin(teams)]
```

```python
output = data[['match_id','batting_team','bowling_team','ball','runs','player_dismissed','city','venue']]
```

```python
output
```

```python
output[output['city'].isnull()]['venue'].value_counts()
```

```python
x = np.where(output['city'].isnull(), output['venue'].str.split().apply(lambda x:x[0]), output['city'])
```

```python
output['city'] = x
```

```python
output.isnull().sum()
```

```python
output.drop(columns=['venue'],inplace=True)
```

```python
total_df = output.groupby('match_id').sum()['runs'].reset_index()
output = output.merge(total_df,on='match_id')
```

```python
output
```

```python
output['current_score'] = output.groupby('match_id')['runs_x'].cumsum()
```

```python
output
```

```python
output['over'] = output['ball'].apply(lambda x:str(x).split(".")[0])
output['ball_no'] = output['ball'].apply(lambda x:str(x).split(".")[1])
```

```python
output['balls_bowled'] = (output['over'].astype('int')*6) + output['ball_no'].astype('int')
```

```python
output['crr'] = round((output['current_score']*6)/output['balls_bowled'],2)
```

```python
output[output['player_dismissed'] != '0']
```

```python
output['player_dismissed'] = output['player_dismissed'].apply(lambda x:0 if x=='0' else 1)
output['player_dismissed'] = output['player_dismissed'].astype('int')
```

```python
output['player_dismissed'] = output.groupby('match_id')['player_dismissed'].cumsum()
output['wickets_left'] = 10 - output['player_dismissed']
```

```python
output[output['player_dismissed'] <11]
```

```python
final_df = output[['match_id','batting_team','bowling_team','runs_x','current_score','balls_bowled','wickets_left','crr','city','runs_y']]
```

```python
final_df
```

```python
final_df = final_df.sample(final_df.shape[0])
```

```python
final_df['balls_left'] = 120 - final_df['balls_bowled']
final_df['balls_left'] = final_df['balls_left'].apply(lambda x:0 if x<0 else x)
```

```python
final_df['crr'] = round((final_df['current_score']*6)/final_df['balls_bowled'],2)
```

```python
final_df
```

```python
final_df.drop(columns=['balls_bowled'],inplace=True)
```

```python
groups = final_df.groupby('match_id')
```

```python
final_df['runs_x'] = pd.to_numeric(final_df['runs_x'], errors='coerce')
```

```python
match_ids = final_df['match_id'].unique()
last_five = []
for id in match_ids:
    match_df = groups.get_group(id)
    rolling_sum = match_df['runs_x'].rolling(window=30).sum()
    last_five.extend(rolling_sum.tolist())
```

```python
final_df['last_five'] = last_five
```

```python
final_df = pd.read_csv("finalData.csv")
```

```python
X = final_df.drop(columns=['match_id','runs_x','runs_y'])
y = final_df['runs_y']
```

```python
X = X[['batting_team', 'bowling_team', 'city', 'current_score', 'balls_left', 'wickets_left', 'crr', 'last_five']]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=1)
```

```python
X_train
```

```python
X_train['bowling_team'].unique_values()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from xgboost import XGBRegressor
from sklearn.metrics import r2_score,mean_absolute_error

trf = ColumnTransformer([
    ('trf',OneHotEncoder(drop='first',handle_unknown='ignore'),['batting_team','bowling_team','city'])
]
,remainder='passthrough')
```

```python
pipe = Pipeline(steps=[
    ('step1',trf),
    ('step2',StandardScaler(with_mean=False)),
    ('step3',XGBRegressor(n_estimators=1000,learning_rate=0.2,max_depth=12,random_state=1))
])
```

```python
pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_test)

print(r2_score(y_test,y_pred))

print(mean_absolute_error(y_test,y_pred))
```
